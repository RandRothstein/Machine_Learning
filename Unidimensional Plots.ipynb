```python
#Univariate Scatter Plot: Plots values of a variable with respect to index
plt.scatter(data.index,data["SepalLengthCm"])
plt.show()
plt.scatter(data["SepalLengthCm"],data["PetalLengthCm"])
plt.show()
```

```python
#Line Plot:Default plot of matplotlib measurement points are ordered 
plt.plot(data["PetalLengthCm"],data["SepalLengthCm"],markevery=20,marker='o')
```

```python
data.shape
```

```python
plt.title('Line plot for Petal Length')
plt.xlabel('index',fontsize=20)
plt.ylabel('Petal Length',fontsize=20)
#plt.plot(data.index,data['PetalLengthCm'],markevery=1,marker='o')
for name,group in data.groupby('Species'):
	plt.plot(group.index,group['PetalLengthCm'],label=name,markevery=1,marker='o')
plt.legend()
plt.show()
```

```python
plt.scatter(data["Species"],data['PetalLengthCm'])
```

```python
plt.title('Line plot for Petal Length')
plt.xlabel('index',fontsize=20)
plt.ylabel('Petal Length',fontsize=20)
plt.plot(data.index,data['PetalLengthCm'],markevery=1,marker='d')
for name,group in data.groupby('Species'):
	plt.plot(group.index,group['PetalLengthCm'],label=name,markevery=1,marker='d')
plt.legend()
```

```python
#Strip Plot
#It also helps to plot the distribution of variables for each category as individual data points.
sns.stripplot(y=data['PetalLengthCm'],x=data['PetalWidthCm'])
plt.xlabel("Index")
plt.ylabel("Species")
plt.legend(labels=("Iris-versicolor","Iris-setosa"))
sns.set(rc={'figure.figsize':(10,10)})
```
